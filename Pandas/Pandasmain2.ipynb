# 7. Data Cleaning In Pandas: -

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\VsCode\AlexTheAnalyst\Pandas\CustomerCallList.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Not_Useful_Column' )
df
```

```python
df['Last_Name'].str.strip()
```

```python
df['Last_Name'] = df['Last_Name'].str.lstrip('...')
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
df['Last_Name'] = df['Last_Name'].str.lstrip('_')
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df
```

```python
df['Last_Name'].str.strip('123._/')
df
```

```python
# df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)
df = df.copy()
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)

# Or df.loc[:, 'Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df
```

```python
df['Phone_Number'].apply(lambda x: x[0:3]+"-"+x[3:6]+"-"+x[6:9])
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
df['Phone_Number'].apply(lambda x: x[0:3]+ "-" +x[3:6]+ "-" + x[6:10])
```

```python
# df["Phone_Number"] = df["Phone_Number"].str.replace("nan--",'')
# df["Phone_Number"] = df["Phone_Number"].str.replace("nan",'')
# df["Phone_Number"] = df["Phone_Number"].str.replace("Na--",'')
# df["Phone_Number"] = df["Phone_Number"].str.replace("Na",'')
df
```

```python
df[["Street Address","State","Zipcode"]] = df['Address'].str.split(",",2, expand = True)

# Future version
# df[["Street Address", "State", "Zipcode"]] = df['Address'].str.split(pat=",", n=2, expand=True)
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
# df['Do_Not_Contact'] = df['Do_Not_Contact'].replace("Yes", "Y")
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace("No", "N")
df
```

```python
# df = df.replace("Na", "")
# df.replace("NaN", "") #it does not replacee bcz its technically empty
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('NNN', 'N')
df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
df
# Another way to drop null value
# df.dropna(subset = 'Phone_Number', inplace = True)
```

```python
df.reset_index()
```

```python
df = df.reset_index(drop = True)
df
```
